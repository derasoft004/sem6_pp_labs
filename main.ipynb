```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('array1.txt', 'r') as f:
    array1 = f.readlines()
array1 = np.array([list(map(int, row.split())) for row in array1])

with open('array2.txt', 'r') as f:
    array2 = f.readlines()
array2 = np.array([list(map(int, row.split())) for row in array2])

with open('result.txt', 'r') as f:
    result = f.readlines()
result = np.array([list(map(int, row.split())) for row in result])
```

```python
np.allclose(array1 @ array2, result)
```

```python
with open('times_seq.txt', 'r') as f:
    data_seq = f.readlines()
with open('times.txt', 'r') as f:
    data_cuda = f.readlines()
```

```python
np.mean(data_seq, axis = 1)
```

```python
data_seq = np.array([list(map(int, row.split())) for row in data_seq])
data_cuda = np.array([list(map(int, row.split())) for row in data_cuda])
```

```python
sizes = np.array([5, 10, 25, 50, 100, 250, 500, 750, 1000, 1100])

fig, ax = plt.subplots()
ax.plot(sizes, np.mean(data_seq, axis = 1) / 1e6, label = 'Среднее, без cuda')
ax.plot(sizes, np.mean(data_cuda, axis = 1) / 1e6, label = 'Среднее, с cuda')
ax.set_xlabel('Размер')
ax.set_ylabel('Время, сек')
ax.set_title('Зависимость времени от размера')
ax.legend()
```

```python
with open('times_cuda2.txt', 'r') as f:
    data_cuda2 = f.readlines()
data_cuda2 = np.array([list(map(int, row.split())) for row in data_cuda2])
```
